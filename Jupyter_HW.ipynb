```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
from fake_useragent import UserAgent
```

```python
title = input("104 search:")
url = f"https://www.104.com.tw/jobs/search/?keyword={title}&order=1&jobsource=2018indexpoc&ro=0"

ua = UserAgent()
user_agent = ua.random
headers = {
    'User-Agent' : user_agent
}

rs = requests.session()
res = rs.get(url, headers=headers)
soup = BeautifulSoup(res.text, 'html.parser')

df = pd.DataFrame(columns=['data_cust_name', 'job_name'])
for i in range(0, len(soup.select('article'))):
    try:
       
        href = soup.select('article')[i].select('a')[0]['href']
        job_code = href.split('/')[-1].split('?')[0]
        url = f'https://www.104.com.tw/job/ajax/content/{job_code}'
        headers = {
            "Accept": "application/json, text/plain, */*",
            "Accept-Encoding": "gzip, deflate, br",
            "Accept-Language": "zh-TW,zh;q=0.9,en-US;q=0.8,en;q=0.7,zh-CN;q=0.6",
            "Connection": "keep-alive",
            "Host": "www.104.com.tw",
            "Referer": f"https://www.104.com.tw/job/{job_code}?jobsource=hotjob_chr",
            "sec-ch-ua": '" Not A;Brand";v="99", "Chromium";v="96", "Google Chrome";v="96"',
            "sec-ch-ua-mobile": "?0",
            "sec-ch-ua-platform": "Windows",
            "Sec-Fetch-Dest": "empty",
            "Sec-Fetch-Mode": "cors",
            "Sec-Fetch-Site": "same-origin",
            "User-Agent": ua.random
        }
        res = rs.get(url, headers=headers)
        res_json = json.loads(res.text)
        
        job_name = soup.select('article')[i]['data-job-name'] # col1
        data_cust_name = soup.select('article')[i]['data-cust-name'] # col2
        job_content = res_json['data']['jobDetail']['jobDescription'] # col3
        
        add_dataframe = pd.DataFrame({'data_cust_name':data_cust_name, 'job_name':job_name, 'content':job_content, "job_code":job_code},index=[1])
        df = df.append(add_dataframe, ignore_index=True)
        
    except:
        pass
    
df
```
