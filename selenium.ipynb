```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np
```

```python
search_input = input()
url = f"https://www.104.com.tw/jobs/search/?keyword={search_input}&order=1&jobsource=2018indexpoc&ro=0"

driver = webdriver.Chrome()
driver.get(url)
```

```python
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
url_list = []
for i in driver.find_elements(By.XPATH, '//*[@id="js-job-content"]/article'):
    new_url = i.find_element(By.XPATH, 'div[1]/h2/a').get_attribute('href')
    url_list.append(new_url)
```

```python
df = pd.DataFrame(columns=['title', 'company', 'content', '職務類別', '工作待遇', '工作性質', '上班地點', '管理責任', '出差外派', '上班時段', '休假制度', '可上班日', '需求人數'])
count = 0
for url in url_list:
    driver.get(url)
    title = driver.find_element(By.CSS_SELECTOR, 'div.job-header__title > h1').get_attribute('title')
    contents = driver.find_element(By.XPATH,'//*[@id="app"]/div[2]/div/div[1]/div[1]/div[2]/div[1]').text
#     contents = driver.find_element(By.XPATH, '//*[@id="app"]/div[2]/div/div[1]/div[1]/div[2]/div[1]')
    company = driver.find_element(By.CSS_SELECTOR, '#app > div.job-header > div:nth-child(2) > div > div > div.job-header__title > div > a:nth-child(1)').get_attribute('title')
    
    length = len(driver.find_elements(By.CSS_SELECTOR, '#app > div.container.jb-container.pt-4.position-relative > div > div.col.main > div.dialog.container-fluid.bg-white.rounded.job-description.mb-4.pt-6.pb-6 > div.job-description-table.row > div' ))
    details = {'title':title, 'company': company, 'content':contents}
    for i in range(2, length+1):
        temp = f'#app > div.container.jb-container.pt-4.position-relative > div > div.col.main > div.dialog.container-fluid.bg-white.rounded.job-description.mb-4.pt-6.pb-6 > div.job-description-table.row > div:nth-child({i})'
       
        detail = driver.find_element(By.CSS_SELECTOR, temp).text.split('\n')
        
        if detail[0] != '':
            details[detail[0]] = str(detail[1:])
            
#     print(title)
#     print(details)
    add_dataframe = pd.DataFrame(details, index=[1])
    df = df.append(add_dataframe, ignore_index=True)
    count = count + 1
    print(f'{count}.{url}')
```
